```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv('13.1 heart-disease.csv')
heart_disease
```

```python
Over_50.head()
```

# Its time to plot using matplotlib object oriented api

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True);
# Putting Data and Customize it with plots
scatter = ax0.scatter(x=Over_50['age'],
                      y=Over_50['chol'],
                      c=Over_50['target']);
ax0.set(title='Checking Heart Disease using Chlestrol Level',
        ylabel='Chlestrol');
# adding legends to graph
ax0.legend(*scatter.legend_elements(), title="Target");
# adding horizontal line
ax0.axhline(y=Over_50['chol'].mean(),
            linestyle='--');

# Putting data to ax1 with age and thalach columns

scatter1 = ax1.scatter(x=Over_50['age'], 
                       y=Over_50['thalach'],
                       c=Over_50['target']);
ax1.set(title='Checking Heart Disease usning thalach',
        xlabel='Age', ylabel='Thalach');

ax1.legend(*scatter1.legend_elements(), title='Target');
ax1.axhline(y=Over_50['thalach'].mean(),
            linestyle='--');

# Adding Title to Figure

fig.suptitle("Heart Disease Analysis by Udemy", fontsize=20, fontweight='bold');

fig.savefig('Heart Disease Analysis by Udemy 1st')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 4 subplots using heart_disease data


```python
plt.style.available
```

```python
plt.style.use('seaborn-whitegrid')

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Setting the data for plot 1
scatter = ax1.scatter(x=heart_disease['age'],
                      y=heart_disease['chol'],
                      c=heart_disease['target'],
                      cmap='winter')
ax1.set(title='Heart Disease and Chlestrol Level',
        xlabel='Age',
        ylabel='Chlestrol')
ax1.legend(*scatter.legend_elements(), title="Legend")
ax1.axhline(heart_disease['chol'].mean(),
            linestyle='--');
# Data to ax2
scatter1 = ax2.scatter(x=heart_disease['age'],
                y=heart_disease['cp'],
                c=heart_disease['target'],
                cmap='summer');
ax2.legend(*scatter.legend_elements(), title='Legend');
ax2.axvline(heart_disease['cp'].mean());
ax2.set(title='Heart Disease and CP Level',
        xlabel='Age', ylabel='CP');
# Data to ax3

scatter2 = ax3.scatter(x=heart_disease['thal'],
                       y=heart_disease['slope'],
                       cmap='winter',
                       c=heart_disease['target']);
ax3.legend(*scatter.legend_elements(), title='Taget')
ax3.axhline(heart_disease['slope'].median(),
            linestyle='--');
ax3.axvline(heart_disease['thal'].mean(),
            linestyle='--');
ax3.set(title='Heart Disease and Thal Level',
        xlabel='Thal',
        ylabel='Slope');
# Setting up the 4th graph
scatter3 = ax4.scatter(x=heart_disease['trestbps'],
                       y=heart_disease['oldpeak'],
                       c=heart_disease['target'],
                       cmap='winter');
ax4.set(title='Heart Disease and Trestbps level and oldpeak',
        xlabel='Trestbps',
        ylabel='Oldpeak');
ax4.legend(*scatter.legend_elements(), title="Target");
ax4.axhline(heart_disease['oldpeak'].mean(), linestyle='--');

# Customize the figure

fig.suptitle('Heart Disease Analysis using multiple factors', fontsize=18, fontstyle='italic');

fig.savefig('Heart Disease Analysis')
```
